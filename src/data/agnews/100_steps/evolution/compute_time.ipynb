```python
import pandas as pd

filename = "results_run_4_seed_2024_propor.csv"
df = pd.read_csv(filename)
    
# Cada geração tem 30 linhas com o mesmo gen_time_min. Usamos o primeiro valor.
gen_times = df.groupby('generation')['gen_time_min'].first()
total_min = gen_times.sum()
total_hours = total_min / 60.0
    
print(f"{filename}: {total_min:.1f} min  ({total_hours:.2f} h)")
```
